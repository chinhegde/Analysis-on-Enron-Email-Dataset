Using the GLOB package to read all files from the folders recursively.
Using the EMAIL package to parse the emails.

```python
import email
import glob
```

```python
files = glob.glob('enron_mail_20110402/maildir/**/**/*.')
```

```python
out = open("emails.txt", "a")
```

Writing the output in emails.txt

```python
for file in files:
  with open(file) as f:
    try:
      lines = f.readlines()
      lines = ''.join(lines)
      e = email.message_from_string(lines)
      out.write(e.get_payload())
    except:
      pass

out.close()
```
